```python
# from __future__ import division, print_function, absolute_import
%matplotlib inline
import numpy as np

from phidl import Device, Layer, LayerSet, make_device
from phidl import quickplot as qp # Rename "quickplot()" to the easier "qp()"
import phidl.geometry as pg
import phidl.routing as pr
import phidl.utilities as pu

# import customed geolib
import photonic as ph
import photonic as ph
import matplotlib.pyplot as plt
```

```python
ww = 0.5


def block(ww=0.1, width=1, length=10):
    D = Device('bus')
    c0 = pg.connector(midpoint=(length*0,0),width=ww,orientation=0)
    c1 = pg.connector(midpoint=(length*.1,0),width=ww,orientation=0)
    c2 = pg.connector(midpoint=(length*.3,width*.5-2*ww),width=ww,orientation=0)
    c3 = pg.connector(midpoint=(length*.5,0),width=ww,orientation=0)
    c4 = pg.connector(midpoint=(length*.7,width*.5-2*ww),width=ww,orientation=0)
    c5 = pg.connector(midpoint=(length*.9,0),width=ww,orientation=0)
    c6 = pg.connector(midpoint=(length,0),width=ww,orientation=0)
    
    D << pr.route_basic(port1=c0.ports[1],port2=c1.ports[2])
    D << pr.route_basic(port1=c1.ports[1],port2=c2.ports[2])
    D << pr.route_basic(port1=c1.ports[1],port2=c2.ports[2])
    D << pr.route_basic(port1=c2.ports[1],port2=c3.ports[2])
    D << pr.route_basic(port1=c3.ports[1],port2=c4.ports[2])
    D << pr.route_basic(port1=c4.ports[1],port2=c5.ports[2])
    D << pr.route_basic(port1=c5.ports[1],port2=c6.ports[2])
    
    D << pg.copy(D).reflect([0,width*.5], [1,width*.5])
    return D.flatten()


D = Device('MZIM')
N=12
M=14
ww=0.1
width=2
length=6
for i in range(N):
    for j in range(M-i%2):
        D << block(ww, width, length).move((i*length,(2*j+i%2)*width))


        
qp(D.flatten())

```
